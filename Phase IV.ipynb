## Exploratory Data Analysis

Given that our dataset consists of over 500 companies over the course of 10 years, we brainstormed ways that we could explore our data to attempt to explain variation observed in our data. Moreover, exploratory data analysis, as shown below, was utilized in order for us to better understand our data.

```python
finalOutput = pd.read_csv('finalOutput.csv')
finalOutput['date']= pd.to_datetime(finalOutput['date'])
finalOutput.describe()
```

Firstly, we analyzed the output of the describe function, which provided us with a glance of the data we were working with. With over 58 thousand observatuons of data, we first noticed outliers. Mainly, that although 50% of our stocks had adjusted close prices between 27 and 80, that there existed companies in the dataset that were much lower in price, as seen with the .76 price, as well as companies much higher in price, such as the 3817 price. While we were intitially concerned that we found a $.76 priced stock in the S&P 500, we realized that this can be explained as this company (ENPH) and similar companies had a very low price at one period of time between 2010-2020, but later joined the S&P 500. We also noticed that there was a very large variation in volume being traded for different stocks. We recognized that this may be due to the fact that some companies are in the news and more well known than other companies. Regarding monthly returns, there are of course extreme variations between company's returns, which could occur due to poor or strong company earnings, bankruptcies, stock splits, and other relevant corporate events. 

```python
finalOutput.boxplot(column = 'monthly_return', by = 'sector')
plt.ylim(-25, 25)
plt.xticks(rotation = 90)
plt.title("Monthly Return (%) by Sector")
plt.ylabel("Monthly Return (%)")
```

Due to the fact that our describe function showed a very large spread in monthly returns in our dataset, we decided to break down monthly return by sector. We identified that the median returns are typically near 1-3%, and that each sector has veyr similar medians. However, one can notice large differences across sectors with regards to the 1st and 4th quartiles. For example, consumer services, energy, and information technology, tend to have the largest ranges, while utility, real estate, financials, and similar less novel sectors have smaller ranges. This can be explained by the volatility that is implied given budding and innovative industries which are suceptible to more market variation and business risk.

```python
colorList=["black","fuchsia","maroon","darkorange","lawngreen","darkslategray","deepskyblue","blueviolet","lightpink","gold","darkkhaki","darkgoldenrod"]
i=0

p = finalOutput.groupby(['sector','date']).mean().reset_index()
for sec in p.sector.unique():
    plt.plot(p[p['sector'] == sec].date, p[p['sector'] == sec].adjclose, label = sec, alpha=.8, color=colorList[i])
    i+=1
    
plt.legend(prop={'size': 13})

plt.title('Monthly Adjusted Closing Price by Sector from 2010 to 2020')

plt.ylabel('Adjusted Closing Price ($)');
```

Utilizing a line chart, we can effectively visualize sector trends over the time period. We can see that these sectors tend to grow over time, with certain sectors growing faster than others. It came to our attention at this point that the market was rebounding from the 2008 Housing Crisis, and this in part can explain the very positive trend expeirenced throughout this visualization. Moreover, we noticed a drop in closing prices in 2019, which can most likely be explained by the COVID pandemic. Additionally, some sectors grew much faster than others, with consumer discretionary, communication services, and healthcare growing at an especially fast rate compared to slower growing sectors such as utilities, energy, and consumer staples. 

```python
finalOutputMonth = finalOutput.copy(deep=False)
finalOutputMonth["month"] = finalOutputMonth['date'].dt.month

finalOutputMonth[finalOutputMonth['ticker'] == "SPY"].boxplot(column="monthly_return", by="month")
plt.title('SPY Monthly Return by Month')
plt.ylabel('Monthly Return (%)')
```

From the previous chart, it is clear that buying and holding stocks across all sectors tends to be profitable. However, our next thought was to determine the best time for entry into the market. By taking the overall S&P 500 index, which represents the performance of all of the 500 companies (weighted) in our dataset, we can try to identify the best times to enter the market. We noticed that most months achieved positive median returns except for August. Also, February and April almost always has positive returns, while months such as January and October tend to have a very large variation in performance. By only analyzing this chart, it can be determined that more speculative and risky traders should purchase the S&P 500 right before months such as October or January, while more conservative traders may want to purchase SPY before less variable months such as February, April, and November.

```python
groupMonth = finalOutputMonth[['sector','month','adjclose', 'monthly_return']].groupby(['sector','month'])
groupMonth.describe().head()
```

```python
meansMonthSec = groupMonth.mean().reset_index()
i=0
for sec in meansMonthSec.sector.unique():
   plt.plot(meansMonthSec[meansMonthSec['sector'] == sec].month, meansMonthSec[meansMonthSec['sector'] == sec].monthly_return, label = sec, alpha=.7,color=colorList[i])
   i+=1
plt.legend(prop={'size': 10})
plt.xticks(np.arange(0, 13, 1));
plt.yticks(np.arange(-4, 6, 1))
plt.title('Average Monthly Return for Each Month')
plt.xlabel('Month')
plt.ylabel('Monthly return (%)');
```

Our previous boxplots that visualized SPY's performance in different months are helpful for swing traders who may the S&P for a month. However, many individual might be investors, that will hold stocks for over a year. In order to provide relevant insights to this audience, we visualized the average performance of each sector over the course of a year, so that investors can try to predict their portfolio's perforamnce. We noticed that over the course of the year, we generally found 3 peaks of highest performance, around the beginning of the year (February), middle of the year (July), and end of the year (October), with 2 low points of negative returns (June and August).

## Dataset Limitations
- We only have data on the largest 500 companies, therefore we might not be able to extrapolate this data to smaller companies
- Regarding the general market, larger companies may have larger covariances (move together) more so than smaller companies, and therefore our analysis might be impacted, and our correlations might not be as telling
- Our dataset only involves 10 years of data, from January 2010 to January 2020. Therefore, we are unable to capture certain large events such as the 2008 housing crisis making our data less relevant in times of economic distress
- Given that our data occurs during one of the greatest bull runs in history, our data might be skewed and not applicable to other time periods
- Our dataset is missing certain fields such as dividends, number of employees, and news events, which might have an impact on the stock price. In other words, we do not have all of the data about each company, and this data might better explain trends in our data.
- Some stocks did not go public until more recently, so we are missing earlier data on them.
- Stocks that were previously part of the S&P 500 but are no longer part of the S&P 500 will not be represented in our data.
