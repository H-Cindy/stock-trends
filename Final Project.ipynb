### Model 1: Do factors like sector and time of year play a role on a stock's monthly performance?

Based on the information learned from our exploratory data analysis, in order to answer this question, we are going to build a model that classifies whether a stock will have above average monthly returns based on its sector, the month, and whether the previous month performed above average.

Creating an 80/20 train/test split

```python
dums = list(outputNew.columns)[13:]
predictors = dums + ['prev_above_avg']
X_train, X_test, y_train, y_test = train_test_split(outputNew[predictors], outputNew['above_avg'], test_size= 0.2)
print('predictor variables: ', predictors)
```

```python
testing_model = LogisticRegression().fit(X_train, y_train)
for var in range(len(predictors)):
    print('Variable: ', predictors[var])
    print('Coefficient:', round(testing_model.coef_[0,var],3))
    print('\n')
print('Intercept: ', round(testing_model.intercept_[0],3))
```

The positive coefficient associated with the Sector "All" shows that the SPY 500 index fund is positively correlated with having an above average month of returns. The sectors Consumer Staples, Energy, Materials, and Real Estate have negative correlations with a stock having an above average month of returns. All other sectors have positive coefficients, and thus are positively correlated with a stock performing above average.

Months 5, 6, 8, 9, and 12 have negative corrrelations with the stock performing above average, while months 1, 2, 3, 4, 7, 10, and 11 have positive correlations.

Finally, based on the negative coefficient -0.191 for "prev_above_avg," a stock performing above average in the previous month indicates that it will be more likely to underperform in the current month.

```python
print(f'Testing model score: {testing_model.score(X_test, y_test):.3f}')
```

```python
scores = cross_val_score(
    LogisticRegression(),
    outputNew[predictors],
    outputNew['above_avg'],
    cv = 20
)
print(f'Mean cross-validated accuracy: {scores.mean():.3f}')
```

```python
print('Baseline accuracy:', round(outputNew.above_avg.sum()/len(outputNew),3))
```

Our model is able to improve classification accuracy by 4% compared to baseline accuracy where we blindly classify every observation as having an above average monthly return.

### Model 2: Do stocks perform better in years after an above average year of percent return (i.e., does one year of strong performance indicate future years of strong performance?)

In order to determine if stocks perform better in the years following an above average year of percent return, we first need to know a company's historical average yearly return for each month, which we calculate below.

```python
def getAverageReturn(ticker,dataframe,date):
    newDataframe=dataframe[(dataframe["ticker"]==ticker) & (dataframe["date"] < date)]
    return newDataframe["monthly_return"].mean() * 12

aboveAverage = pd.read_csv('finalOutput.csv')
aboveAverage['date']= pd.to_datetime(aboveAverage['date'])
aboveAverage["average_yearly"] = aboveAverage.apply(lambda x: getAverageReturn(x['ticker'],aboveAverage,x["date"]), axis=1)
```

Next, we create a new column, which determines if a stock's performance is above the company's annual average. To make this process easier, we can determine the annual performance for the stock. Once we do this we can easily calculate if there is an above average year or not.

```python
def get_yearly(dataframe):
    i=0
    while i < dataframe.shape[0]:
        startingPrice=dataframe.iloc[i]["adjclose"]
        startingTicker=dataframe.iloc[i]["ticker"]
        startingYear=dataframe.iloc[i]["date"].year

        i+=1
        #Keep going until change in ticker or year
        while i < dataframe.shape[0] and dataframe.iloc[i]["date"].year == startingYear and dataframe.iloc[i]["ticker"] == startingTicker:
            i+=1
        
        #Need to go back one
        i-=1
        
        dataframe.at[i,"yearly_return"] =  (dataframe.iloc[i]["adjclose"] - startingPrice)/startingPrice * 100
        i+=1
        
    return dataframe
    
get_yearly(aboveAverage)
aboveAverage["yearly_return"]=None
aboveAverage= get_yearly(aboveAverage)
aboveAverage.to_excel("aboveAverage.xlsx") 
```

Now that we have the average yearly return for a stock as well as the stocks yearly return, we can perform the analysis. We can first filter the dataframe to remove all yearly_return with 0 or None, so that we effectively filter for all the valid full years from January to December.

```python
yearlyPerformance=aboveAverage[aboveAverage["yearly_return"].notnull() & aboveAverage["yearly_return"]!=0]
yearlyPerformance.reset_index(inplace=True)
```

Now we can create a new column that determines if a year's performance is above average.

```python
def isAboveAverage(avgRet,yearlyRet):
    if yearlyRet > avgRet:
        return 1
    return 0

yearlyPerformance["is_above_avg"] = yearlyPerformance.apply(lambda x: isAboveAverage(x["average_yearly"],x['yearly_return']), axis=1)
yearlyPerformance.to_excel("yearlyPerformance.xlsx") 
```

Finally, to conduct our analysis, we can use a logistic regression to determine if a stock's above average performance in one year can predict an above average performance in the next 5 years. We decided to use 5 years time, as this is typically what is used for in many areas of finance for extrapolation.

```python
def getNextFive(row,dataframe):
    
    startingTicker=dataframe.iloc[row]["ticker"]
    startingYear=dataframe.iloc[row]["date"].year
    if startingYear > 2014:
        return None
    
    count = 0

    for i in range(1,6):
        if startingTicker!=dataframe.iloc[row+i]["ticker"]:
            return None
        if dataframe.iloc[row+i]["is_above_avg"] == True:
            count +=1
    return count / 5


yearlyPerformance["next_5_above_avg_probability"] = yearlyPerformance.apply(lambda x: getNextFive(x.name,yearlyPerformance), axis=1)
yearlyPerformance.to_excel("yearlyPerformance.xlsx") 
```

```python
yearlyPerformanceFiltered = yearlyPerformance[yearlyPerformance["next_5_above_avg_probability"].notnull()]
hot_hand_model = LogisticRegression().fit(yearlyPerformanceFiltered[['next_5_above_avg_probability']], yearlyPerformanceFiltered['is_above_avg'])
print("Logistic Regression Coefficient:",round(hot_hand_model.coef_[0][0],3))
print("Logistic Regression Intercept:",round(hot_hand_model.intercept_[0],3))
```

```python
plt.scatter(
    yearlyPerformanceFiltered["next_5_above_avg_probability"], 
    yearlyPerformanceFiltered["is_above_avg"], 
    alpha = 0.3,
    label='Above-average year'
)

#  logit probability of class 1
plt.scatter(
    yearlyPerformanceFiltered["next_5_above_avg_probability"], 
    hot_hand_model.predict_proba(yearlyPerformanceFiltered[['next_5_above_avg_probability']])[:,1], 
    alpha = 0.3,
    label='Predicted probability of\nabove-average future performance over 5 years'
)
plt.xlabel("next_5_above_avg_probability")
plt.ylabel("is_above_avg")

plt.legend()
plt.show()
```

We can determine from this regression that there is a negative relationship between a stock having an above average year, and the number of above average years that will ensue over the next 5 years, given a model coefficient -0.284. Given a very slightly negative intercept of -0.016, this demonstrates that along with a negative correlation, that a stock is always more likely to have an average or below average year than an above average year. For example, if the next 5 years of a stock underperforms the average return, the given stock is still more likely to underperform in the current year.

### Model 2 - Evaluation of Significance

We will now perform a hypothesis test with p values and a confidence interval to determine the significance of the relationship of our model.

```python
def permute(input_array):

    permuted = input_array.copy().values 
    np.random.shuffle(permuted)
    return pd.Series(permuted) 

def plot_model_line(df, model, is_resampled=True):
    if is_resampled:
        color="grey"
        alpha=0.1
    else:
        color="steelblue"
        alpha=0.7
    plt.xlabel("Expected Years of Above Average Performance Over 5 years")
    plt.ylabel("Current Year Above Average Performance")
    plt.plot(df['next_5_above_avg_probability'], df['next_5_above_avg_probability'] * model.coef_[0] + model.intercept_, color=color, alpha=alpha)

for i in range(100):
    permuted_model = LogisticRegression().fit(yearlyPerformanceFiltered[['next_5_above_avg_probability']], permute(yearlyPerformanceFiltered['is_above_avg']))
    plot_model_line(yearlyPerformanceFiltered, permuted_model)

model = LogisticRegression().fit(yearlyPerformanceFiltered[['next_5_above_avg_probability']], yearlyPerformanceFiltered['is_above_avg'])
plot_model_line(yearlyPerformanceFiltered, model, is_resampled=False)

plt.show()
```

```python
steeper_slopes = 0
n_permutations = 500
permuted_slopes = np.zeros(n_permutations) 

for i in range(n_permutations):
    permuted_model = LogisticRegression().fit(yearlyPerformanceFiltered[['next_5_above_avg_probability']], permute(yearlyPerformanceFiltered['is_above_avg']))
    permuted_slopes[i] = permuted_model.coef_[0]
    if np.abs(permuted_model.coef_[0]) > np.abs(model.coef_):
        steeper_slopes += 1
    
print(f'Fraction of permuted slopes more extreme than observed slopes: {steeper_slopes / n_permutations}')
```

At a p-value of .05, given a simulated p-value of .132, the negative relationship that we observe between a year's above average performance and the proportion of the next 5 years that have an above average performance is not significant. 

```python
n_bootstrap = 500
bootstrap_slopes = np.zeros(n_bootstrap)
for i in range(n_bootstrap):
    resampled_data = yearlyPerformanceFiltered.sample(n=len(yearlyPerformanceFiltered), replace=True)
    resampled_model = LogisticRegression().fit(resampled_data[['next_5_above_avg_probability']], resampled_data['is_above_avg'])
    bootstrap_slopes[i] = resampled_model.coef_[0]

print("Confidence Interval: ",np.quantile(bootstrap_slopes, q=[0.025, 0.975]))
```

Given the confidence interval and p-value, we fail to reject the null hypothesis that there is no relationship between an above average year and the amount of above average years over the next 5 years, is present within the interval. Therefore, we cannot choose to accept that there is not a significant relationship between a stock's above average performance in a year and the number of above average performances that occur in the next 5 years.

### Model 2
Interestingly enough, we in fact calculated a negative slope between a stock's above average yearly performance and its yearly performance over the next 5 years, although this could have been due to random chance as shown by our confidence interval, and a simulated p-value of 0.132. However, through the use of a logistic regression, confidence interval, and hypothesis test, there fails to show a significant *positive* relationship between a stock's above average yearly performance and its yearly performance over the next 5 years. Therefore, our analysis contradicts the "hot hand" theory, as it seems that if a stock has an above average performance that there is no strong evidence suggesting a positive relationship with the performance of the stock over the next 5 years, and perhaps even a negative relationship. 

It is questionable, though, if this finding can be applied to time periods other than from 2010 - 2020, as over this time period the market experienced one of the largest bull runs in history. Further, since we were working with the S&P 500, since these stocks are more stable, it is understandable that these companies may move differently than smaller cap stocks. Therefore, this behavior may not necessarily apply to smaller cap stocks, which may grow at increasingly fast rates as they expand and gain market share. Therefore, we cannot confidently say that the "hot hand" does not apply to non S&P 500 companies on the market. 

From this analysis, we determined that although we cannot reject the null hypothesis, that there appears to be a negative relationship between a stock having an above average year, and the number of above average years that will occur over the next 5 years, given a model coefficient -0.284. Further, we found that with a model intercept of -0.016, that a stock is always more likely to have an average or below average year than an above average year, regardless of the number of above average years a stock has over the next 5 years. Therefore, when we retroactively analyze a stock's performance in a given year, it will always be more likely to have an average or underperformance regardless of its perforamnce over the next 5 years.

Despite an inability to apply these findings to smaller cap stocks, disproving the “hot hand” for the S&P 500 over yearly intervals is an incredibly important finding for active mutual funds, which currently invest trillions of dollars into indexes, such as the S&P 500. The concept of momentum trading, which looks to capitalize on the inertia of a price trend to continue after rising or falling, is widely used in active investing and trading. However, it is clear from our findings that this strategy on the S&P 500 did not outperform the market from 2010 - 2020.


## Dataset Limitations

- Stocks that were previously part of the S&P 500 but are no longer part of the S&P 500 will not be represented in our data.
